```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Step 1: Load the image
image = Image.open('./predicted_masks.png')  # Load the image (ensure it is in PNG format to preserve colors)
image_np = np.array(image)  # Convert image to NumPy array

# Step 2: Get unique colors and count pixels for each
unique_colors, counts = np.unique(image_np.reshape(-1, image_np.shape[-1]), axis=0, return_counts=True)

# Step 3: Display the unique colors and the pixel count for each color
for color, count in zip(unique_colors, counts):
    print(f"Color {color} has {count} pixels")

# Optional: Visualize the image
plt.imshow(image_np)
plt.title("Image with Different Colors")
plt.show()
```

```python
from PIL import Image

# Step 1: Load the image
image = Image.open('image3.jpg')  # Replace with your image path

# Step 2: Resize the image to 1024x1024
image_resized = image.resize((1024, 1024))

# Step 3: Break the image into four 512x512 parts
# Coordinates for the four quadrants
coords = [
    (0, 0, 512, 512),         # Top-left
    (512, 0, 1024, 512),      # Top-right
    (0, 512, 512, 1024),      # Bottom-left
    (512, 512, 1024, 1024)    # Bottom-right
]

# Step 4: Save each part
for i, coord in enumerate(coords):
    cropped_image = image_resized.crop(coord)
    cropped_image.save(f'part_{i+1}.jpg')  # Save each 512x512 part
```

```python
from PIL import Image

# Step 1: Load the four 512x512 images
image_1 = Image.open('pp1.jpg')  # Top-left
image_2 = Image.open('pp2.jpg')  # Top-right
image_3 = Image.open('pp3.jpg')  # Bottom-left
image_4 = Image.open('pp4.jpg')  # Bottom-right

# Step 2: Create a new blank image (1024x1024)
final_image = Image.new('RGB', (1024, 1024))

# Step 3: Paste each image into its respective quadrant
final_image.paste(image_1, (0, 0))           # Top-left
final_image.paste(image_2, (512, 0))         # Top-right
final_image.paste(image_3, (0, 512))         # Bottom-left
final_image.paste(image_4, (512, 512))       # Bottom-right

# Step 4: Save the final joined image
final_image.save('joined_image.jpg')  # Replace with desired output path
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Step 1: Load the two images
image1 = Image.open('./joined_image.jpg')
image2 = Image.open('./result3.jpg')

# Step 2: Ensure both images are the same size (resize if necessary)
image1 = image1.resize((1024, 1024))  # Adjust size if needed
image2 = image2.resize((1024, 1024))  # Adjust size if needed

# Step 3: Convert images to NumPy arrays
image1_np = np.array(image1)
image2_np = np.array(image2)

# Step 4: Plot and overlay images with adjustable opacity
fig, ax = plt.subplots()

# Display the first image
ax.imshow(image1_np)

# Overlay the second image with adjustable opacity
ax.imshow(image2_np, alpha=0.7)  # Set the opacity level (0 to 1)

# Step 5: Hide axis and show the result
plt.axis('off')
plt.show()

# Optional: Save the overlay image
fig.savefig('overlay_image.png', bbox_inches='tight', pad_inches=0)
```
